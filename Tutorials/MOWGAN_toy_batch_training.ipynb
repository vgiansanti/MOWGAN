### Load your data

```python
data1 = sc.datasets.blobs(n_variables=1100, n_centers=3, cluster_std=1.0, n_observations=7000)
data2 = sc.datasets.blobs(n_variables=2100, n_centers=3, cluster_std=.7, n_observations=8000)
data3 = sc.datasets.blobs(n_variables=100, n_centers=3, cluster_std=.5, n_observations=1000)
```

```python
sc.pp.pca(data1)
sc.pp.pca(data2)
sc.pp.pca(data3)
```

```python
sc.pl.pca(data1)
sc.pl.pca(data2)
sc.pl.pca(data3)
```

```python
sc.pp.neighbors(data1)
sc.pp.neighbors(data2)
sc.pp.neighbors(data3)
```

```python
sc.tl.umap(data1)
sc.tl.umap(data2)
sc.tl.umap(data3)
```

```python
sc.tl.leiden(data1)
sc.tl.leiden(data2)
sc.tl.leiden(data3)
```

```python
sc.pl.umap(data1, color='leiden')
sc.pl.umap(data2, color='leiden')
sc.pl.umap(data3, color='leiden')
```
